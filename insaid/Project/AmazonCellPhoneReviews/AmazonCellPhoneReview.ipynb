```python
# How many mobiles are released till date per year
reviews.groupby(reviews.date.dt.year).count()
```

```python
reviews['year'] = reviews.date.dt.year
reviews['month'] = reviews.date.dt.month
reviews['week'] = reviews.date.dt.week
reviews['day'] = reviews.date.dt.day
```

**Remove the unused columns**

```python
reviews.drop(labels=['title', 'body'], axis=1, inplace=True)
```

**How many Mobile Models got reviewed**

```python
reviews.asin.nunique()
```

**How many models are released in each year**

```python
reviews.groupby(['year'])['asin'].nunique()
```

```python
reviews.groupby(['asin', 'rating']).count().reset_index()
```

```python
reviews_rated = pd.get_dummies(data=reviews, columns=['rating'], dtype='int').groupby('asin').sum().reset_index()
reviews_rated.drop(labels=['verified', 'helpfulVotes', 'year', 'month', 'week', 'day'], axis=1, inplace=True)
```

**Consolidate Ratings per brand**

```python
#reviews_rated['totalReviews'] = reviews_rated['rating_1'] + reviews_rated['rating_2'] + reviews_rated['rating_3'] + reviews_rated['rating_4'] + reviews_rated['rating_5']
reviews_rated['totalReviews'] = reviews_rated[['rating_1', 'rating_2', 'rating_3', 'rating_4', 'rating_5']].sum(axis=1)
reviews_rated.head()
#print(reviews_rated.totalReviews[reviews_rated.asin == 'B004H23JXW']) # supposed to be 348
```

**Display the mobile reviews in sorting order by total reviews**

```python
reviews_rated.sort_values(by = 'totalReviews', ascending=False).head(15)
```
