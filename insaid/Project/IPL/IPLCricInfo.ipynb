## 1. Problem Statement

The **Indian Premier League (IPL)** is a professional Twenty20 cricket league in India contested during March or April and May of every year by eight teams representing eight different cities in India. The league was founded by the Board of Control for Cricket in India (BCCI) in 2008. IPL has an exclusive window in ICC Future Tours Programme.

The IPL is the most-attended cricket league in the world and in 2014 ranked sixth by average attendance among all sports leagues. In 2010, the IPL became the first sporting event in the world to be broadcast live on YouTube. The **brand value of IPL in 2019 was Rs 47,500 crore (US$6.7 billion)**, according to Duff & Phelps. According to BCCI, **the 2015 IPL season contributed Rs 11.5 billion to the GDP of the Indian economy**.

IPL is a platform for emerging players who are looking for a chances in International Cricket, where young team members are emerging from IPL.

#### A general function to plot a graph

```python
def plotGraph(xValues, yValues, hoverText, xAxisTitle, yAxisTitle, graphTitle):
    #hoverText = np.around(np.array(yValues)/sum(yValues)*100, decimals=1)
    fig = go.Figure(
        data = [
            go.Bar(
                x = xValues,
                y = yValues,
                text = yValues,
                textposition='auto',
                hovertext = hoverText
            )
        ]
    )
    
    fig.update_layout(
        title = graphTitle,
        xaxis_tickfont_size = 14,
        xaxis_tickangle = -45,
        xaxis_title = xAxisTitle,
        yaxis=dict(
            title = yAxisTitle,
            titlefont_size = 16,
            tickfont_size = 14,
        ),
        legend=dict(
            x = 0,
            y = 1.0,
            bgcolor = 'rgba(255, 255, 255, 0)',
            bordercolor = 'rgba(255, 255, 255, 0)',
        ),
        barmode = 'group',
        bargap = 0.15, # gap between bars of adjacent location coordinates.
        bargroupgap = 0.1 # gap between bars of the same location coordinate.
    )
    
    fig.show()
```

```python
from ipywidgets import widgets

def batsmanStats(df, batsmanName):
    #seasonsPlayed = df[(df.batsman == batsmanName)]['season'].nunique()
    matchesPlayed = df[(df.batsman == batsmanName)]['match_id'].nunique()
    ballsFaced = df[(df.batsman == batsmanName)]['ball'].count()
    runsScored = df[(df.batsman == batsmanName)]['batsman_runs'].sum()
    dismissals = df[(df.player_dismissed == batsmanName)]['dismissal_kind'].count()
    runOuts = df[(df.batsman != batsmanName) & (df.player_dismissed == batsmanName) & (df.dismissal_kind == 'run out')]['dismissal_kind'].count()
    foursHit = df[(df.batsman == batsmanName) & (df.batsman_runs == 4)]['batsman_runs'].count()
    sixesHit = df[(df.batsman == batsmanName) & (df.batsman_runs == 6)]['batsman_runs'].count()
    ballsDefended = df[(df.batsman == batsmanName) & (df.batsman_runs == 0)]['batsman_runs'].count()
    ducks = df[df.batsman == batsmanName][['match_id', 'batsman_runs']].groupby(['match_id']).sum()['batsman_runs'].eq(0).value_counts().values[1]
    notDismissed = matchesPlayed - dismissals
    strikeRate = runsScored / ballsFaced * 100
    average = runsScored/matchesPlayed

    stats = dict()
    stats['playerName'] = batsmanName
    stats['matchesPlayed'] = matchesPlayed
    stats['ballsFaced'] = ballsFaced
    stats['runsScored'] = runsScored
    stats['dismissals'] = dismissals
    stats['runOuts'] = runOuts
    stats['foursHit'] = foursHit
    stats['sixesHit'] = sixesHit
    stats['ballsDefended'] = ballsDefended
    stats['ducks'] = ducks
    stats['notDismissed'] = notDismissed
    stats['strikeRate'] = round(strikeRate, 2)
    stats['average'] = round(average, 2)
    
    #print(stats)

    df = pd.DataFrame([stats]).set_index('playerName')
    #print(df)
    df.iplot(kind='bar')

# Drop Down List
w = widgets.Dropdown(
    options=top10BatsMen,
    value=top10BatsMen[5],
    description='Top Most Batsman: '
)

def on_change(change):
    if change['type'] == 'change' and change['name'] == 'value':
        #print ("changed to %s" % change['new'])
        batsmanName = change['new']
        batsmanStats(deliveries, batsmanName)
    
    push_notebook()

w.observe(on_change)
display(w)
```
