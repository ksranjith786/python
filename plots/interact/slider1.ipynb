```python
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

```python
def f(x):
    plt.plot(np.arange(0, 10), x * np.arange(0, 10))
    plt.ylim(-40, 40)
```

```python
f(5)
```

```python
interact(f, x = 5)
```

```python
interact(f, x=(-10, 10, 0.5))
```
