```python
import pandas as pd
df = pd.read_csv('http://bit.ly/drinksbycountry')
df
```

```python
df.beer_servings.mean()
```

```python
df.groupby('continent').mean()
```

```python
df.groupby('continent').wine_servings.max()
```

```python
df.groupby('continent').wine_servings.min()
```

```python
df.groupby('continent').wine_servings.count()
```

```python
df.groupby('continent').wine_servings.agg(['count', 'min', 'max', 'mean', 'median'])
```

```python
%matplotlib inline
df.groupby('continent').mean().plot()
```
